```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import re
```

```python
mypath = '02_Coreference_Resolution/leader_resolved'
files_list = [f for f in listdir(mypath) if isfile(join(mypath, f))]

long_leader_df = pd.DataFrame([])
print(files_list)
```

```python
for file in files_list:
    temp_df = pd.read_csv('{}/{}'.format(mypath, file))
    leader_col = re.match('(.*)(?=_resolve)', file)[1]
    temp_df = temp_df.drop([leader_col], axis=1)
    temp_df['leader_name'] = leader_col
    long_leader_df = pd.concat([long_leader_df, temp_df], ignore_index=True)

print('done')
```

```python
long_leader_df.sample(n=15)

```

```python
print(len(long_leader_df))

long_leader_df.to_csv('all_leaders_resolved.csv')
```
